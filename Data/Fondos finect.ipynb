```python
import re
import requests
from bs4 import BeautifulSoup
import re
import time
import pandas as pd
import re, time, pandas as pd, requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# urls = [
#     "https://www.finect.com/fondos-inversion/FR0000447823-Axa_tresor_court_terme_c",
#     "https://www.finect.com/fondos-inversion/FR0000989626-Groupama_tresorerie_ic",
#     "https://www.finect.com/fondos-inversion/FR0000991390-La_francaise_tresorerie_isr_r",
#     "https://www.finect.com/fondos-inversion/FR0010288316-Groupama_entreprises_n",
#     "https://www.finect.com/fondos-inversion/LU0083138064-Bnp_paribas_euro_money_market_c_cap",
#     "https://www.finect.com/fondos-inversion/LU0366536711-Pictetsovereign_shortterm_mm_eur_p",
#     "https://www.finect.com/fondos-inversion/LU0034353002-Dws_floating_rate_notes_lc",
#     "https://www.finect.com/fondos-inversion/IE00BG226Z29-Aegon_european_abs_a_eur_acc",
#     "https://www.finect.com/fondos-inversion/FI0008811997-Evli_nordic_corporate_bond_b",
#     "https://www.finect.com/fondos-inversion/IE00BFZMJT78-Neuberger_berman_ultr_st_er_bd_eur_iacc",
#     "https://www.finect.com/fondos-inversion/ES0112618006-Bh_deuda_fi",
#     "https://www.finect.com/fondos-inversion/LU1706854152-Amundi_sf_diversified_st_bd_a_eur_nd",
#     "https://www.finect.com/fondos-inversion/LU0840158819-Storm_bond_rc_eur",
#     "https://www.finect.com/fondos-inversion/IE00BD2B9710-Twelve_cat_bond_p_eur_acc",
#     "https://www.finect.com/fondos-inversion/LU2023201044-Pareto_nordic_cross_credit_h_i_eur",
#     "https://www.finect.com/fondos-inversion/ES0124880032-Cs_renta_fija_05_b_fi",
#     "https://www.finect.com/fondos-inversion/FR0000027609-Lazard_euro_short_duration_sri_ic",
#     "https://www.finect.com/fondos-inversion/LU1694212348-Nordea_1__low_dur_eurp_cov_bd_bp_eur",
#     "https://www.finect.com/fondos-inversion/LU1670718219-Mg_lux_short_dtd_corp_bd_a_eur_acc",
#     "https://www.finect.com/fondos-inversion/IE00BZ005D22-Aegon_european_abs_b_eur_acc",
#     "https://www.finect.com/fondos-inversion/FI0008800511-Evli_short_corporate_bond_b",
#     "https://www.finect.com/fondos-inversion/FR0010149120-Carmignac_securite_aw_eur_acc",
#     "https://www.finect.com/fondos-inversion/IE00BH65QP47-Vanguard_glb_st_bd_idx__h_acc",
#     "https://www.finect.com/fondos-inversion/IE00BDFB7290-Vanguard_glb_st_corp_bd_idx__h_acc",
#     "https://www.finect.com/fondos-inversion/ES0140072028-Abaco_renta_fija_mixta_global_c_fi",
#     "https://www.finect.com/fondos-inversion/ES0140794001-Gamma_global_fi",
#     "https://www.finect.com/fondos-inversion/LU1623762843-Carmignac_pf_credit_a_eur_acc",
#     "https://www.finect.com/fondos-inversion/LU0942882589-Sigma_inv_house_fcp_brightg_glb_inc_a_i",
#     "https://www.finect.com/fondos-inversion/IE00BDTYYP61-Man_glg_hi_yld_opps_d_eur_acc",
#     "https://www.finect.com/fondos-inversion/LU1670722161-Mg_lux_glb_fl_rt_hy_a_h_eur_acc",
#     "https://www.finect.com/fondos-inversion/IE00B67T5G21-Ishares_euro_invgrdcorpbdidxieinstacc",
#     "https://www.finect.com/fondos-inversion/IE0007471471-Vanguard_us_govt_bd_idx__h_acc",
#     "https://www.finect.com/fondos-inversion/IE00B18GC888-Vanguard_global_bd_idx_eur_h_acc",
#     "https://www.finect.com/fondos-inversion/IE000E4XZ7U3-Man_glg_dynamic_income_i_h_eur_cap",
#     "https://www.finect.com/fondos-inversion/IE000JM1MJ59-Man_glg_euro_corporate_bond_d_eur_acc",
#     "https://www.finect.com/fondos-inversion/LU1988110927-Buy__hold_luxembourg_bh_bond_class_1",
#     "https://www.finect.com/fondos-inversion/IE00B296X584-Aegon_high_yield_global_bd_a_acc_eur_hdg",
#     "https://www.finect.com/fondos-inversion/IE00B296YK09-Aegon_strategic_global_bd_a_acc_eur_hdg",
#     "https://www.finect.com/fondos-inversion/LU2482630675-Mg_total_return_crdt_invmt_p_eur_acc",
#     "https://www.finect.com/fondos-inversion/FR0010235507-Lazard_credit_opportunities_pc_eur",
#     "https://www.finect.com/fondos-inversion/LU0571066462-Vontobel_global_high_yld_bd_b_eur",
#     "https://www.finect.com/fondos-inversion/LU2416422751-Vontobel_credit_opps_hi_hdg_eur_cap",
#     "https://www.finect.com/fondos-inversion/IE00B246KL88-Vanguard_20_yr__trs_idx__acc",
#     "https://www.finect.com/fondos-inversion/IE00B03HD191-Vanguard_glb_stk_idx__acc",
#     "https://www.finect.com/fondos-inversion/IE00BD0NCM55-Ishares_dev_wld_idx_ie_d_acc_eur",
#     "https://www.finect.com/fondos-inversion/IE00BYX5NX33-Fidelity_msci_world_index_eur_p_acc",
#     "https://www.finect.com/fondos-inversion/IE0032126645-Vanguard_us_500_stk_idx__acc",
#     "https://www.finect.com/fondos-inversion/IE0031786696-Vanguard_em_mkts_stk_idx__acc",
#     "https://www.finect.com/fondos-inversion/LU0302296495-Dnb_fund_technology_retail_a",
#     "https://www.finect.com/fondos-inversion/LU1213836080-Fidelity_global_technology_aacceur",
#     "https://www.finect.com/fondos-inversion/IE00BDGV0290-Guinness_global_equity_income_d_eur_acc",
#     "https://www.finect.com/fondos-inversion/IE00BH481053-Gqg_partners_global_equity_a_eur_acc",
#     "https://www.finect.com/fondos-inversion/LU0625737910-Pictetchina_index_p_eur",
#     "https://www.finect.com/fondos-inversion/LU0329070915-Jupiter_india_select_l_eur_acc",
#     "https://www.finect.com/fondos-inversion/IE00B42W4L06-Vanguard_glb_smallcp_idx__acc",
#     "https://www.finect.com/fondos-inversion/LU1598719752-Cobas_international_p_acc_eur",
#     "https://www.finect.com/fondos-inversion/LU1598720172-Cobas_large_cap_p_acc_eur",
#     "https://www.finect.com/fondos-inversion/ES0141116030-Hamco_global_value_fund_r_fi",
#     "https://www.finect.com/fondos-inversion/IE00BH6XSF26-Heptagon_kopernik_glb_allcp_eq_ae__acc",
#     "https://www.finect.com/fondos-inversion/ES0112611001-Azvalor_internacional_fi",
#     "https://www.finect.com/fondos-inversion/ES0146309002-Horos_value_internacional_fi",
#     "https://www.finect.com/fondos-inversion/LU1984712320-Janus_henderson_hrzn_glb_sc_a2_eur_acc",
#     "https://www.finect.com/fondos-inversion/ES0165243025-Myinvestor_value_c_fi",
#     "https://www.finect.com/fondos-inversion/LU1953238794-Ec_sicav_evercapital_invms_ucits_i_rtl",
#     "https://www.finect.com/fondos-inversion/ES0112231008-Avantage_fund_a_fi",
#     "http://www.finect.com/fondos-inversion/ES0175437005-Dunas_valor_prudente_r_fi",
#     "https://www.finect.com/fondos-inversion/ES0175414012-Dunas_valor_equilibrado_r_fi",
#     "https://www.finect.com/fondos-inversion/ES0175316019-Dunas_valor_flexible_r_fi",
#     "https://www.finect.com/fondos-inversion/LU1694789451-Dnca_invest_alpha_bonds_a_eur",
#     "https://www.finect.com/fondos-inversion/LU1694789535-Dnca_invest_alpha_bonds_b_eur",
#     "https://www.finect.com/fondos-inversion/ES0157935000-Acacia_renta_dinamica_fi",
#     "https://www.finect.com/fondos-inversion/LU1112771768-Helium_fund_selection_seur",
#     "https://www.finect.com/fondos-inversion/IE00BLP5S460-Jupitermerian_glb_eq_absrt_l__h_acc",
#     "https://www.finect.com/fondos-inversion/LU1593227850-Bsf_asia_pacific_divers_eq_ar_d2_eur_h",
#     "https://www.finect.com/fondos-inversion/LU2210152745-Fidelity_abs_ret_glb_eq_ypfacceur_h",
#     "https://www.finect.com/fondos-inversion/LU1164219682-Axawf_euro_credit_total_ret_a_cap_eur",
#     "https://www.finect.com/fondos-inversion/IE00B3Q8M574-Gam_star_cat_bond_ord_eur_acc",
#     "https://www.finect.com/fondos-inversion/ES0168997015-Penta_inversion_b_fi",
#     "https://www.finect.com/fondos-inversion/FR0011042134-Scor_sustainable_euro_loans_c_eur"
# ]

urls = [
    "https://www.finect.com/fondos-inversion/LU0034353002-Dws_floating_rate_notes_lc",
    "http://finect.com/fondos-inversion/ES0175437005-Dunas_valor_prudente_r_fi",
    "https://www.finect.com/fondos-inversion/ES0138922069-Gesconsult_corto_plazo_i_fi",
    "https://www.finect.com/fondos-inversion/ES0173319031-Renta_4_renta_fija_euro_fi",
    "https://www.finect.com/fondos-inversion/IE00BG226Z29-Aegon_european_abs_a_eur_acc",
    "https://www.finect.com/fondos-inversion/FI0008811997-Evli_nordic_corporate_bond_b",
    "https://www.finect.com/fondos-inversion/IE00BFZMJT78-Neuberger_berman_ultr_st_er_bd_eur_iacc",
    "https://www.finect.com/fondos-inversion/ES0112618006-Bh_deuda_fi",
    "https://www.finect.com/fondos-inversion/LU1706854152-Amundi_sf_diversified_st_bd_a_eur_nd",
    "https://www.finect.com/fondos-inversion/LU2200514128-Pareto_nordic_cross_credit_r_eur",
    "https://www.finect.com/fondos-inversion/LU0840158819-Storm_bond_rc_eur",
    "https://www.finect.com/fondos-inversion/ES0140794001-Gamma_global_fi",
    "https://www.finect.com/fondos-inversion/LU1670722161-Mg_lux_glb_fl_rt_hy_a_h_eur_acc",
    "https://www.finect.com/fondos-inversion/LU1988110927-Buy__hold_luxembourg_bh_bond_class_1",
    "https://www.finect.com/fondos-inversion/LU2222028099-Merchbanc_fcp_renta_fija_flex_a",
    "https://www.finect.com/fondos-inversion/LU0942882589-Sigma_inv_house_fcp_brightg_glb_inc_a_i",
    "https://www.finect.com/fondos-inversion/LU0170291933-Candriam_bds_global_hi_yld_c_eur_cap",
    "https://www.finect.com/fondos-inversion/LU1623762843-Carmignac_pf_credit_a_eur_acc",
    "https://www.finect.com/fondos-inversion/ES0140072028-Abaco_renta_fija_mixta_global_c_fi",
    "https://www.finect.com/fondos-inversion/IE00BDTYYP61-Man_glg_hi_yld_opps_d_eur_acc",
    "https://www.finect.com/fondos-inversion/FR0013506987-Lazard_euro_sht_dur_high_yld_sri_rc_eur",
    "https://www.finect.com/fondos-inversion/ES0116567035-Cartesio_x_fi",
    "https://www.finect.com/fondos-inversion/IE00BD2B9710-Twelve_cat_bond_p_eur_acc",
    "https://www.finect.com/fondos-inversion/IE00B246KL88-Vanguard_20_yr__trs_idx__acc",
    "https://www.finect.com/fondos-inversion/IE00B03HD191-Vanguard_glb_stk_idx__acc",
    "https://www.finect.com/fondos-inversion/IE0031786696-Vanguard_em_mkts_stk_idx__acc",
    "https://www.finect.com/fondos-inversion/LU0625737910-Pictetchina_index_p_eur",
    "https://www.finect.com/fondos-inversion/LU1213836080-Fidelity_global_technology_aacceur",
    "https://www.finect.com/fondos-inversion/LU1598719752-Cobas_international_p_acc_eur",
    "https://www.finect.com/fondos-inversion/LU3038481936-Hamco_global_value_r_eur_acc",
    "https://www.finect.com/fondos-inversion/ES0146309002-Horos_value_internacional_fi",
    "https://www.finect.com/fondos-inversion/ES0112611001-Azvalor_internacional_fi",
    "https://www.finect.com/fondos-inversion/ES0175414012-Dunas_valor_equilibrado_r_fi",
    "https://www.finect.com/fondos-inversion/ES0175316019-Dunas_valor_flexible_r_fi",
    "https://www.finect.com/fondos-inversion/LU1694789451-Dnca_invest_alpha_bonds_a_eur",
    "https://www.finect.com/fondos-inversion/LU1112771503-Helium_fund_selection_beur",
    "https://www.finect.com/fondos-inversion/IE00BLP5S460-Jupitermerian_glb_eq_absrt_l__h_acc",
    "https://www.finect.com/fondos-inversion/LU1001747408-Jpm_europe_eq_abs_alp_a_perf_acc_eur",
    "https://www.finect.com/fondos-inversion/LU1508158430-Bsf_asia_pacific_divers_eq_ar_a2_eur_h"
]

HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/126.0.0.0 Safari/537.36"
    ),
    "Accept-Language": "es-ES,es;q=0.9,en;q=0.8",
}

def _clean_text(s: str) -> str:
    import re
    return re.sub(r"\s+", " ", (s or "").strip())

def _normalize_key(s: str) -> str:
    import re
    s = s.strip().lower()
    s = s.replace("€", "eur").replace("%", "pct").replace("/", "_").replace("-", " ")
    s = re.sub(r"[^\w\s]", "", s)
    s = re.sub(r"\s+", "_", s)
    return s

def _to_number(txt: str):
    import re
    raw = _clean_text(txt)
    if not raw:
        return None
    is_pct = "%" in raw
    num = re.sub(r"[^\d,.\-+]", "", raw)
    try:
        if "," in num and "." in num:
            if num.rfind(",") > num.rfind("."):
                val = float(num.replace(".", "").replace(",", "."))
            else:
                val = float(num.replace(",", ""))
        elif "," in num:
            val = float(num.replace(".", "").replace(",", "."))
        else:
            val = float(num)
        return val/100.0 if is_pct else val
    except ValueError:
        return raw

def _parse_ratios_from_container(container):
    """Parsea filas de ratios dentro de un contenedor de pestaña (pane)."""
    ratios = {}
    row_re   = re.compile(r"^partials__RowBlock", re.I)
    name_re  = re.compile(r"^partials__HeaderColumn", re.I)
    value_re = re.compile(r"Label-sc", re.I)

    for row in container.find_all("div", class_=row_re):
        name_div = row.find("div", class_=name_re)
        if not name_div:
            continue
        key = _clean_text(name_div.get_text(" "))
        if not key:
            continue

        cols = row.find_all("div", class_=re.compile(r"^partials__Column", re.I))
        right_div = cols[1] if len(cols) >= 2 else row

        label_span = right_div.find("span", class_=value_re) or right_div.find("span")
        if label_span:
            inner_val = label_span.find("span")
            val_txt = _clean_text((inner_val or label_span).get_text(" "))
        else:
            val_txt = _clean_text(right_div.get_text(" "))

        if not val_txt:
            continue

        ratios[f"ratios__{_normalize_key(key)}"] = _to_number(val_txt)

    return ratios

def scrape_finect_ratios_tabindex(url: str, tab_index: int = 0) -> dict:
    """
    Extrae ratios de la pestaña indicada por índice:
      0 = suele ser '12 meses' (por defecto)
      1 = '3 años' (típicamente)
      2 = '5 años'
      ... (depende de la página)
    """
    r = requests.get(url, headers=HEADERS, timeout=25)
    r.raise_for_status()
    soup = BeautifulSoup(r.text, "lxml")

    # Todos los panes de pestañas
    panes = soup.find_all("div", class_=re.compile(r"^TabChild-sc-", re.I))
    if not panes:
        # Fallback: intenta parsear todo el documento (quizás no hay tabs)
        return _parse_ratios_from_container(soup)

    if tab_index < 0 or tab_index >= len(panes):
        raise IndexError(f"tab_index fuera de rango. Hay {len(panes)} panes detectados.")

    target_pane = panes[tab_index]
    return _parse_ratios_from_container(target_pane)

def extract_valor_liquidativo(soup):
    s = soup.select_one('div[class^="partials__Inner-sc-1bigw9a-0"] span[class^="Label-sc-1esfagn-0"] span')
    return _to_number(s.get_text(strip=True)) if s else None



def build_ratios_table_custom(urls, sleep_sec=0.5):


    HEADERS = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/126.0.0.0 Safari/537.36",
        "Accept-Language": "es-ES,es;q=0.9,en;q=0.8",
    }

    # ✅ Nuevo mapeo de sufijos
    SUFFIX_MAP = {
        2: "",
        3: "__12m",
        4: "__3y",
        5: "__5y",
        6: ""
    }
    PERIOD_ORDER = ["", "__12m", "__3y", "__5y"]

    def _clean_text(s: str) -> str:
        return re.sub(r"\s+", " ", (s or "").strip())

    def _get_benchmark(url: str, timeout: int = 20):
        try:
            r = requests.get(url, headers=HEADERS, timeout=timeout)
            r.raise_for_status()
            soup = BeautifulSoup(r.text, "lxml")
            text = soup.get_text(" ", strip=True)
            for pat in [
                r"Benchmark\s*[:\-]\s*([^\n\r|·]+)",
                r"Índice\s+de\s+referencia\s*[:\-]\s*([^\n\r|·]+)",
                r"Índice\s*[:\-]\s*([^\n\r|·]+)",
            ]:
                m = re.search(pat, text, flags=re.I)
                if m:
                    return _clean_text(m.group(1))
        except Exception:
            pass
        return None

    rows = []
    for i, u in enumerate(urls, 1):
        base = {"url": u}
        m = re.search(r"/fondos-inversion/([A-Z0-9]{12})", u, re.I)
        base["isin"] = m.group(1) if m else None
        base["benchmark"] = _get_benchmark(u)

        row = dict(base)
        r = requests.get(u, headers=HEADERS, timeout=25)
        soup = BeautifulSoup(r.text, "lxml")
        row["valor_liquidativo"] = extract_valor_liquidativo(soup)
        try:
            for idx in range(7):
                try:
                    ratios = scrape_finect_ratios_tabindex(u, tab_index=idx)
                    for k, v in ratios.items():
                        # pestañas 0 y 1: sin sufijo
                        if idx in (0, 1):
                            final_key = k
                        else:
                            suf = SUFFIX_MAP.get(idx, f"__t{idx}")
                            final_key = f"{k}{suf}"
                        row[final_key] = v
                except IndexError:
                    continue
                except Exception as e:
                    row[f"error_tab_{idx}"] = str(e)
                    continue

            rows.append(row)
            print(f"[{i}/{len(urls)}] OK {row.get('isin') or ''}")
        except Exception as e:
            print(f"[{i}/{len(urls)}] ERROR {u}: {e}")
            row["error"] = str(e)
            rows.append(row)

        time.sleep(sleep_sec)

    df = pd.DataFrame(rows)

    # Selecciona las métricas principales
    def col_exists(name): return name in df.columns

    returns_cols, vol_cols, sharpe_cols, mdd_cols = [], [], [], []

    for suf in PERIOD_ORDER:
        c = f"ratios__rentabilidad_anualizada{suf}"
        if col_exists(c): returns_cols.append(c)

    for suf in PERIOD_ORDER:
        c = f"ratios__volatilidad{suf}"
        if col_exists(c): vol_cols.append(c)

    for suf in PERIOD_ORDER:
        c1 = f"ratios__ratio_de_sharpe{suf}"
        c2 = f"ratios__sharpe{suf}"
        if col_exists(c1):
            sharpe_cols.append(c1)
        elif col_exists(c2):
            sharpe_cols.append(c2)

    for suf in PERIOD_ORDER:
        c = f"ratios__maxima_caida{suf}"
        if col_exists(c): mdd_cols.append(c)

    meta_cols = [c for c in ["isin", "benchmark"] if c in df.columns]
    trailing_cols = [c for c in ["url"] if c in df.columns]
    error_cols = sorted([c for c in df.columns if c.startswith("error")])

    selected = set(meta_cols + returns_cols + vol_cols + sharpe_cols + mdd_cols + trailing_cols + error_cols)
    other_cols = [c for c in df.columns if c not in selected]

    ordered_cols = meta_cols + returns_cols + vol_cols + sharpe_cols + mdd_cols + trailing_cols + other_cols + error_cols

    return df[ordered_cols]


df_all = build_ratios_table_custom(urls, sleep_sec=0.5)

def select_and_order_columns(df):
    # 1️⃣ Eliminar prefijo "ratios__" de todos los nombres
    df = df.rename(columns={c: c.replace("ratios__", "") for c in df.columns})

    # 2️⃣ Eliminar columna fecha_de_inicio si existe
    if "fecha_de_inicio" in df.columns:
        df = df.drop(columns=["fecha_de_inicio"], errors="ignore")

    # 3️⃣ Crear columna auxiliar "nombre_fondo" a partir del enlace URL
    if "url" in df.columns:
        df["nombre_fondo"] = (
            df["url"]
            .str.extract(r"/[A-Z0-9]{12}-(.+)$")[0]  # captura lo que hay tras el ISIN-
            .str.replace("_", " ", regex=False)     # reemplaza _ por espacio
            .str.title()                            # formato bonito: Axa Tresor Court Terme C
        )

    # 4️⃣ Lista definitiva de columnas y orden deseado (añadimos nombre_fondo)
    final_cols = [
        "nombre_fondo",
        "gestora",
        "isin",
        "categoría",
        "benchmark",
        "rentabilidad__12m",
        "rentabilidad_anualizada__3y",
        "rentabilidad_anualizada__5y",
        "volatilidad__12m",
        "volatilidad__3y",
        "volatilidad__5y",
        "ratio_de_sharpe__12m",
        "ratio_de_sharpe__3y",
        "ratio_de_sharpe__5y",
        "máxima_caída__12m",
        "máxima_caída__3y",
        "máxima_caída__5y",
        "valor_liquidativo",
        "comisión_de_gestión",
        "gastos_corrientes",
        "comisión_de_éxito",
        "comisión_de_custodia",
        "comisión_de_suscripción",
        "comisión_de_reembolso",

    ]

    # 5️⃣ Eliminar columnas completamente vacías
    df = df.dropna(axis=1, how="all")
    df = df.loc[:, ~(df.replace("", pd.NA).isna().all())]

    # 6️⃣ Mantener solo las que existen realmente
    final_cols_existing = [c for c in final_cols if c in df.columns]

    # 7️⃣ Devolver el DataFrame limpio y ordenado
    return df[final_cols_existing]

df_all = select_and_order_columns(df_all)
# output_file = "C:/Users/Rubén/Dropbox/FIN/Data/fondos_selected.xlsx"
# output_file = "C:/Users/estudis/Dropbox/FIN/Data/fondos_selected.xlsx"
df_all.to_excel(output_file, index=False)

```
